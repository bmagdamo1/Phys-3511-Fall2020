```python
import numpy as np
import matplotlib.pyplot as plt
#for inline plots
%matplotlib notebook
```

```python
def Do_MCpi(N):
    points = np.random.rand(N,2)
    #get size of the points
    norm = np.linalg.norm(points,axis=1) #treats each row as a vector and normailizes it
    
    #check if its in the unit circle
    inside_circle = norm <= 1 #bool
    #print(inside_circle)
    
    #get x-y of points in circle for plotting: their x-y values
    cpoints = points[norm<=1]
    #print(cpoints)
    
    number_inside_circle = np.sum(inside_circle)
    print('pi = ',4*number_inside_circle/float(N))

    ##note in general it is better to explicitly force our sum and N to be floats
    #print(type(N))
    #print(type(number_inside_circle))
    
    #Make the plot
    fig = plt.figure('N = {}'.format(N,0))
    plt.plot(points[:,0],points[:,1],'ko',label = 'Outside')
    plt.plot(cpoints[:,0],cpoints[:,1],'ro', label = 'Inside')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.tight_layout()
```

```python
Do_MCpi(10)
```
